#### ColumnTransformer----->  Column transformer lets us  apply different transformations to different columns of our dataset.
#### scikit-learn(sklearn)----> sklearn is a popular python library used for machine learning.
#### preprocessing----> preprocessing is a module inside sklearn that contain tool to prepare data before training a machine larning model
####                   this includes tools to scale data,encode labels,normalize etc.

#### OneHotEncoder --->  OneHotEncoder converts categorical (text) values into numerical form (binary dummy variables).

#### "encoder"----->	A name for this transformation (can be anything).

#### OneHotEncoder(drop=None)---->	Apply a one-hot encoding transformation. The (drop=None) parameter means don’t drop any columns(i.e.include all dummy variables).

#### (remainder="passthrough") ----->	This means: keep all other columns unchanged in the output.

#### df=ct.fit_transform(df) ------> fit_transform() both fits the encoder and transforms the data.

#### It returns a NumPy array, not a pandas DataFrame.
#### So df here becomes a NumPy array after this line.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer([("encoder",OneHotEncoder(drop=None),[-4])],remainder="passthrough")
df=ct.fit_transform(df)
df
```

#### StandardScaler---> It is a feature scaling technique from sklearn.preprocessing that standardizes numerical data

#### It transforms each columns so that it has:
#### 1) Mean = 0
#### 2) Standard deviation = 1
#### This is called standardization.

#### df = scaler.fit_transform(df)--->
#### 1) fit() — computes the mean and standard deviation for each column in df.
#### 2) transform() — uses those values to scale the data.
#### So all columns in df are standardized.
#### df----> The result is a NumPy array of scaled values (not a DataFrame anymore).

####  use of scaling--->
####  1) Many ML algorithms (like, Logistic Regression, KNN) work better or converge faster when all features are on the same scale.
#### 2) It prevents features with large numeric ranges from dominating smaller ones.

#### 2) Testing data---> It is used to evalute how well the model perform on unseen data.

#### Logistic regression----> Predict the probability that an input belongs to a certain class (eg. 0 or 1).
####                         It uses the logistic (sigmoid) function to squeeze any realvalued number into a range between 0 and 1.


```python
# logistic regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

#### classifier.fit(X_train, y_train)---->Train the model on the training data.

```python
classifier.fit(x_train,y_train)
```

#### classifier.predict(X_test)----> Predict class labels (0 or 1) for new data.

```python
y_pred=classifier.predict(x_test)
```

#### y_pred------> Predicted output values of  model obtained by passing test data (x_test) to our trained model.

```python
y_pred
```

#### Accuracy score for LogisticRegression

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

#### Train the DecisionTreeClassifier
  



```python
from sklearn.tree import DecisionTreeClassifier
```

#### Decision tree classifier:It is a supervised machine learning algorithm used for classification tasks where the goal is to predict categories or labels.
#### It works by splitting data into smaller and smaller groups based on feature values, forming a tree-like structure — just like a flowchart!

#### Working of this code--->
####  Creates a decision tree that:
#### Can grow to a depth of 5 splits.
#### Will only split a node if it contains at least 3 samples.
#### This makes the tree more controlled and less likely to overfit.

```python
classifier=DecisionTreeClassifier(max_depth=3,min_samples_split=3)
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

#### KNeighborsClassifier : KNeighborsClassifier is a machine learning algorithm used for classification problems.
